- Data encoding : it is an way to convert any catogrical data into numerical data .

: here since we are clear from begning that machine learning is all about learning pattern from data .mean some sort of mathmatical calculation but for that data has to be numrical . as far as i know our mosty data is categorical
1. so first step is going to be transform this categorical data in numrical data using data encoding >>

```python
# 1. begaining from medal columns
df['Medal']
```

: 1 . here since medal columns is not in order therfor i am converting it into numerical which gives order to our texual data .

: label encoder : It gives numrical data to each categorial data by follwing an certain order .
- here to medal columns it will asign an order value to our each prsent medal such as gold repersented by 2 , silver = 1 and bronze = 0 .
- it says gold medal ios highest medal or more signfincat then silver which is more repective then bronze
- it will help us to find which athlete will win which kind of medal like gold , silver or bronze


```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
label_encoder  # here our label encoder is empty

# here i have filled data into label encoder of medal column . henceforht it has been given numrical order
encoded_medal = label_encoder.fit_transform(df['Medal'])

# now in order to concatinvate and given header i converted this column into dataframe
encoded_medal = pd.DataFrame(encoded_medal, columns =['encoded_medal'])

# now concatinating with original data
df = pd.concat([df,encoded_medal],axis =1)
#now whole medal columns has been converted into numrical here stored as encoded medal
```

2. Second target categorical columns is gender . since it is in order therefor i will be using one hot encoding or in other words i say nominal encoding
- what does it do -- it create dummy variables for based on the parsent data in target column . here we are having two values it means two dummy varibles will be created .

```python
df['Gender']
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder()
one_hot_encoder

#HERE just filling data into one hot incoder
encoded_gender = one_hot_encoder.fit_transform(pd.DataFrame(df['Gender'])).toarray()

# here i changed the new tranformed data into dataset and have given them header fot batter understending
encoded_gender = pd.DataFrame(encoded_gender,columns = ['Male',"Female"])
# here one repersents yes and 0 mens NO


# here in anohter words in male columns if zero means not male which repersent to femals and if one means male .
# only one columns can repserent our gender categroy therfor i dont need both columns
# so i removed here female orinted dummy varible column to keep our data cleaned .
encoded_gender.drop("Female",axis =1 , inplace = True)

encoded_gender.columns = ['is_athlete_Male'] # here i changed columns header name in order to more batter understanding

# here concatinated the above gender col with orignal dataframe
df = pd.concat([df,encoded_gender],axis =1 )
```

3. Now changing sports categorical column into numrical column . since i know sports have multiple columns and not having any order .
- therfor i will use here one hot encoding to convert it into numerical  


```python
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder2 = OneHotEncoder() # here our OHE is empty

encoded_sports = one_hot_encoder2.fit_transform(pd.DataFrame(df['Sport'])).toarray()

# finding the all features names
one_hot_encoder2.get_feature_names_out()

# now turning this normal data into data frame and giving header to it
encoded_sports = pd.DataFrame(encoded_sports, columns =one_hot_encoder2.get_feature_names_out())

# now cancating it with the original data
df = pd.concat([encoded_sports,df], axis = 1)
```

## model selection and its performace test >>

1. Model : random forest classifier()
- in order to get more accurate information using this model rather then logistics regression

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# definig into varible
random_forest = RandomForestClassifier(random_state = 42)

# now feeding the all trin data in order to train our model
random_forest.fit(x_train, y_train)

# here making perdiction over medal by serving features test data
y_prdict_RFC = random_forest.predict(x_test)

# now checking our model performance
accuracy1 = accuracy_score(y_test,y_prdict_RFC)
cm1 = confusion_matrix(y_test,y_prdict_RFC)
report = classification_report(y_test,y_prdict_RFC)

print("perdiction : ",y_prdict_RFC)
print("accuracy : ",accuracy1)
print("confusion metrix :",cm1)
print('report', report)
```
